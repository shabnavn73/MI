```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scaled_data=scalar.fit_transform(data)
```

```python
scaled_data
```

```python
scaled_data=pd.DataFrame(scaled_data,columns=data.columns)
```

```python
scaled_data.head
```

```python
model = KMeans(n_clusters=3,
               init='k-means++',
               n_init=10,
               max_iter=300,
               tol=0.0001,
               verbose=0,
               random_state=42,
               copy_x=True,
               algorithm='auto')

model.fit(scaled_data)
model.inertia_
```

```python
clusters = range(1, 20)
sse=[]
for cluster in clusters:
    model = KMeans(n_clusters=cluster,
               init='k-means++',
               n_init=10,
               max_iter=300,
               tol=0.0001,
               verbose=0,
               random_state=42,
               copy_x=True,
               algorithm='auto')

    model.fit(scaled_data)
    sse.append(model.inertia_)

sse_df = pd.DataFrame(np.column_stack((clusters, sse)), columns=['cluster', 'SSE'])
fig, ax = plt.subplots(figsize=(13, 5))
ax.plot(sse_df['cluster'], sse_df['SSE'], marker='o')
ax.set_xlabel('Number of clusters')
ax.set_ylabel('Inertia or SSE')
```

```python
model = KMeans(n_clusters=5,
               init='k-means++',
               n_init=10,
               max_iter=300,
               tol=0.0001,
               verbose=0,
               random_state=42,
               copy_x=True,
               algorithm='auto')

model.fit(scaled_data)
```

```python
print('SSE: ', model.inertia_)
print('\nCentroids: \n', model.cluster_centers_)

pred = model.predict(scaled_data)
data['cluster'] = pred
print('\nCount in each cluster: \n', data['cluster'].value_counts())
```
